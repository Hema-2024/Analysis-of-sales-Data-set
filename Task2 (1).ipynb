```python
d["status"] = d["status"].fillna("Nan")
```

```python
d["stars"]
```

```python
d["stars"] = d["stars"].str.replace(" star rating","")
```

```python
d["stars"] = d["stars"].astype("float64")
```

```python
d.isnull().sum()
```

```python
d.stars.value_counts()
```

```python
def reviewRank(tl):
    if tl > 2:
        return 1
    else:
        return 0
    
```

```python
d["reviewRank"] = d.stars.apply(reviewRank)
```

```python
def enter1():
    labels = ["Reviewed","Not Reviewed"]
    plt.pie(d.status.value_counts(), labels = labels,autopct='%.0f%%')
    plt.show()
```

```python
labels=["Positive","Negative"]
plt.pie(d.reviewRank.value_counts(),labels=labels,autopct='%.0f%%')
plt.show()
```
